```python
from queue import PriorityQueue

file = open('Input file.txt','r')
info = file.readlines()
file.close()
legend = {'Arad':'A','Neamt':'F','Bucharest':'Z','Oradea':'B','Craiova':'S','Pitesti':'P','Eforie':'T','RimnicuVilcea':'R','Fagaras':'O','Timisoara':'C','Dobreta':'V','Urziceni':'D','Hirsova':'N','Vaslui':'H','Iasi':'Q','Zerind':'E','Lugoj':'G','Mehadia':'L','Sibiu':'Si','Giurgiu':'Gi'}


#removing \n and assigning legends#
for i in range(len(info)):
    info[i]=info[i][0:-1]
    
    for item in legend.items():
        if item[0] in info[i]:
            info[i] = info[i].replace(item[0], item[1])
#---------------------------------------------------------#


#seperating heuristic values#
heur_values={}
for i in info:
    i = i.split(' ')
    heur_values[i[0]]=int(i[1])
#---------------------------------------------------------#



#creating graph#
graph = {}
for i in info:
    i=i.split(' ')
    i.pop(1)
    graph[i[0]]=[]
    for j in range(1,len(i),2):
        graph[i[0]].append((i[j],int(i[j+1])))
#---------------------------------------------------------#        


#----------------------A* Algorithm----------------------#
q = PriorityQueue()
visited = []
path = []
path_found = False


#INPUT#
strt = input('Enter starting place:').lower().capitalize()
dest =  input('Enter destination:').lower().capitalize()
name_list = list(legend.keys())
letter_list = list(legend.values())
strt = letter_list[name_list.index(strt)]
dest = letter_list[name_list.index(dest)]


q.put((heur_values[strt],strt,None)) #(heur,node,prev_parent)

while q.empty()==False:
    
    
    parent=q.get() #(f_n,legend,prev_parent)
    
    if parent[1] not in path:
        visited.append(parent[1])
        
    if parent[2] not in path:
        path.append(parent[2]) 
        
    if parent[1]==dest:
        path_found = True
        path.append(dest)
        break

    for children in graph[parent[1]]: #children = ('node',weight)
        
        if children[0] not in visited:
            f_n = parent[0]-heur_values[parent[1]]+children[1]+heur_values[children[0]]
            q.put((f_n,children[0],parent[1]))
            
#---------------------------------------------------------#                
            
if path_found:
    
    #calculating total distance
    path.pop(0)
    total_distance = 0
    for i in range(len(path)):
        for j in graph[path[i]]:
            if i<len(path)-1:
                if j[0]==path[i+1]:
                    total_distance+=j[1]
    
    
    #replacing legend with actual names
    name_list = list(legend.keys())
    letter_list = list(legend.values())
    updated_path = []
    for i in path:
        updated_path.append(name_list[letter_list.index(i)])
        
    
    #printing final information
    print('Path: ',end='')
    for i in range(0,len(updated_path)-1):
        print(updated_path[i],end='->')                  
    print(updated_path[-1])
    print('Total distance:',total_distance,'km')
                               

else:
    print('NO PATH FOUND')
    
    
    
            
        
        
        
   

  
    
```
